```python
import numpy as np
import math
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as ss
import scipy.optimize as so
import statistics as st

#1#
#t#
n = 20
k = 20

def t_power_Norm(mu):
  pow = []
  for i in range(n):
    p = []
    for j in range(n):
      X_1 = ss.norm.rvs(0, 1, size = k)
      X_2 = ss.norm.rvs(mu, 1, size = k)
      res = ss.ttest_ind(X_1, X_2)
      p.append(res.pvalue)
    p=np.sort(p)
    pow.append(p[0])
  pow = np.sort(pow)
  return pow

def t_power_Lap(mu):
  pow = []
  for i in range(n):
    p = []
    for j in range(n):
      X_1 = ss.laplace.rvs(0, 1, size = k)
      X_2 = ss.laplace.rvs(mu, 2, size = k)
      res = ss.ttest_ind(X_1, X_2)
      p.append(res.pvalue)
    p=np.sort(p)
    pow.append(p[0])
  pow = np.sort(pow)
  return pow

plt.title("t power norm")
ax=np.arange(0, 1, 1/n)
t_pow_n1 = t_power_Norm(0)
t_pow_n2 = t_power_Norm(0.1)
t_pow_n3 = t_power_Norm(0.2)
t_pow_n4 = t_power_Norm(0.3)
plt.plot(t_pow_n1, ax)
plt.plot(t_pow_n2, ax)
plt.plot(t_pow_n3, ax)
plt.plot(t_pow_n4, ax)
plt.legend (('mu = 0', 'mu = 0.1', 'mu = 0.2', 'mu = 0.3'))
plt.show()

plt.title("t power lap")
t_pow_l1 = t_power_Lap(0)
t_pow_l2 = t_power_Lap(0.1)
t_pow_l3 = t_power_Lap(0.2)
t_pow_l4 = t_power_Lap(0.3)
plt.plot(t_pow_l1, ax)
plt.plot(t_pow_l2, ax)
plt.plot(t_pow_l3, ax)
plt.plot(t_pow_l4, ax)
plt.legend (('mu = 0', 'mu = 0.1', 'mu = 0.2', 'mu = 0.3'))
plt.show()
```

```python
#  
n = 100
def power_t(mu):
  m=0
  for j in range(n):
    X_1 = ss.norm.rvs(0, 1, size = k)
    X_2 = ss.norm.rvs(mu, 1, size = k)
    res = ss.ttest_ind(X_1, X_2)
    if(abs(res.statistic)<=2):
      m+=1
  return 1-m/n

def power_t_lap(mu):
  m=0
  for j in range(n):
    X_1 = ss.laplace.rvs(0, 1, size = k)
    X_2 = ss.laplace.rvs(mu, 2, size = k)
    res = ss.ttest_ind(X_1, X_2)
    #print(res.statistic/(np.sqrt((k**4)/6)))
    if(abs(res.statistic)<=2):
      m+=1
  return 1-m/n

mu = np.arange(0, 1, 0.01)
t = []
t_lap = []
for i in range(len(mu)):
  t.append(power_t(mu[i]))
  t_lap.append(power_t_lap(mu[i]))
t = np.sort(t)
t_lap = np.sort(t_lap)

plt.title("power / expectation")
plt.plot(mu, t)
plt.plot(mu, t_lap)
plt.legend (('norm', 'lap'))
```
