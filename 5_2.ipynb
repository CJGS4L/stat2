```python
#3*#

import matplotlib.pyplot as plt
import numpy as np
from collections import deque


def nod(a, b):
  while b:
    a, b = b, a % b
  return int(a)

def nod_array(arr):
  nod_value = arr[0]
  for elem in arr:
    nod_value = nod(nod_value, elem)
  return nod_value

def read_matrix():
  n = len(lines)
  print(n)
  print()
  P = np.zeros((n, n))
  for i, line in enumerate(lines):
    row = [float(prob) for prob in line.strip().split(';')]
    P[i, :] = row
  return P

def find_classes(P):
  n = P.shape[0]
  visited = np.zeros(n)
  classes = []

  def dfs(node, current_class):
    visited[node] = 1
    current_class.append(node)
    for neighbor in range(n):
      if P[node, neighbor] > 0 and visited[neighbor] == 0:
        dfs(neighbor, current_class)

  for i in range(n):
    if visited[i] == 0:
      current_class = []
      dfs(i, current_class)
      classes.append(current_class)
  return classes


def find_lengthes(P, class_elem):
  lengthes = []
  n = P.shape[0]
  visited = np.zeros(n)
  q = deque()
  q.append(class_elem[0])
  visited[class_elem[0]] = 1

  while q:
    node = q.popleft()
    level = visited[node]
    for neighbor in class_elem:
      if P[node, neighbor] > 0 and visited[neighbor] == 0:
        visited[neighbor] = level + 1
        q.append(neighbor)
      elif P[node, neighbor] > 0 and visited[neighbor] > 0: # то есть вернулись в какую-то вершину
        length = level + 1 - visited[neighbor]
        lengthes.append(length)
        break
  return lengthes


def find_subclasses(P, class_elem, d:int):
  n = P.shape[0]
  visited = np.zeros(n)
  subclasses = []

  def dfs(node):
    level = visited[node] % d + 1
    for neighbor in class_elem:
      if P[node, neighbor] > 0 and visited[neighbor] == 0:
        visited[neighbor] = level
        dfs(neighbor)

  visited[class_elem[0]] = 1
  dfs(class_elem[0])

  for i in range(1, d + 1):
    indices = np.where(visited == i)[0]
    subclasses.append(list(indices))
  return subclasses


P = read_matrix()
classes = find_classes(P)
for elem in classes:
  d = nod_array(find_lengthes(P, elem))
  print("Class:", elem)
  print("Period =", d)
  print("Subclasses:", find_subclasses(P, elem, d))
  print()
```
